```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
columns = [
    "Architecture",
    "Development cost",
    "Sourcing cost",
    "E-BOM cost",
    "Manufacturing cost",
    "Transport and handling cost",
    "Service and maintenance cost"
]
data = [
    ["A",5,10,20,15,30,15],
    ["B",10,20,15,30,10,10],
    ["C",15,15,25,15,20,25],
    ["D",10,10,10,30,15,30]
]

# Create the pandas DataFrame
df = pd.DataFrame(data, columns = columns)
  
# print dataframe.
df
```

```python
df.set_index('Architecture').plot(kind='bar', stacked=True).legend(loc='center left',bbox_to_anchor=(1.0, 0.5))
```

```python
df.set_index('Architecture').plot(kind="bar", legend=False, subplots=True, layout=(2, 3), figsize=(6, 6), sharex=False)
```
